# Step 3: Cleaning the data.

```python
#checking how many null values we have in each column
data.isnull().sum()
```

```python
#Since the columns REMARKS has the maximum null values so we will be dropping this entire column. Also this entire column doesn't have any effect in our dataset
data.drop(['Remarks'],axis=1,inplace=True)
```

```python
data.columns
```

# After a thorough checking, I found that few things in the dataset needs to get replaced else that will effect my analysis. Thus doing the necessary steps

```python
data['AmountInUSD']=data['AmountInUSD'].apply(lambda x:float(str(x).replace(",","")))
```

```python
data['Date']
```

```python
data['Date']=data['Date'].replace({"12/05.2015":"12/05/2015"})
data['Date']=data['Date'].replace({"13/04.2015":"13/04/2015"})
data['Date']=data['Date'].replace({"15/01.2015":"15/01/2015"})
data['Date']=data['Date'].replace({"22/01//2015":"22/01/2015"})
data['StartupName']=data['StartupName'].replace({'Ola Cabs':'Ola'})
data['StartupName']=data['StartupName'].replace({'Flipkart.com':'Flipkart'})
data['StartupName']=data['StartupName'].replace({'Olacabs':'Ola'})
data['StartupName']=data['StartupName'].replace({'Couponmachine.in':'Couponmachine'})
data['StartupName']=data['StartupName'].replace({'practo':'Practo'})
data['InvestmentType']=data['InvestmentType'].replace({'PrivateEquity':'Private Equity'})
data['InvestmentType']=data['InvestmentType'].replace({'SeedFunding':'Seed Funding'})
data['IndustryVertical']=data['IndustryVertical'].replace({'Ecommerce':'eCommerce'})
data['IndustryVertical']=data['IndustryVertical'].replace({'ECommerce':'eCommerce'})
data['IndustryVertical']=data['IndustryVertical'].replace({'ecommerce':'eCommerce'})
```

# question 7: how many fundings were made year-month wise?

```python
data['monthwise']=(pd.to_datetime(data['Date'],format ='%d/%m/%Y').dt.year*100)+(pd.to_datetime(data['Date'],format='%d/%m/%Y').dt.month)
a=data['monthwise'].value_counts().sort_values(ascending=False)
a
```

```python
year_monthd=data['monthwise'].value_counts()
```

```python
plt.figure(figsize=(15,8))
sns.barplot(a.index,a.values,alpha=0.9)
plt.xticks(rotation='vertical')
plt.xlabel('month and year wise transaction', fontsize=20)
plt.ylabel('No. of fundings made',fontsize=20)
plt.show()
```

# Conclusion: July and August of 2015 had the highest investment, maybe because of the digital India campaign.

# January of 2016 sees the highest investment of 2016 , maybe because of the Startup India Initiative

# We see a decrease in investment from November 2016 to July 2017, maybe because of the demonitization
