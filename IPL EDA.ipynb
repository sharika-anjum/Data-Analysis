```python
#function to find a batsman kis team k against sabse zyada score kiye hain
def batsman_vs_team(batsman):
    
    mask=delivery['batsman']==batsman
    vk=delivery[mask]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
    
    
```

```python
batsman_vs_team('SK Raina')
```

```python
#kis bowler ne sabse zyada run kiye against which team?
def team_vs_bowler(bowler):
    mask=delivery['bowler']==bowler
    v=delivery[mask]
    return v.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
team_vs_bowler('M Ashwin')
```

```python
delivery=pd.read_csv('deliveries.csv')
```

```python
#to find who is the most dangerous batsman in the death over
m=delivery['over']>15
mask=delivery[m]
```

```python
new1=mask.groupby('batsman')['batsman_runs'].sum().reset_index()
new1.sort_values('batsman_runs',ascending=False,inplace=True)
```

```python
new2=mask.groupby('batsman')['ball'].count().reset_index()
new2.sort_values('ball',ascending=False,inplace=True)
```

```python
new3=new1
new3['ball']=new2['ball']
new3['strike_rate']=((new3['batsman_runs']/new3['ball'])*100).astype(float)
new3[new3.ball >200].sort_values('strike_rate',ascending=False).head()
```

```python
#sabse economical bowler
do=delivery[delivery['over']>15]
a=do.groupby('bowler')['total_runs'].count()  #no. of balls bowled by a bowler per offer
b=do.groupby('bowler')['total_runs'].count()>100
bowler=a[b].index.tolist()
newdf=delivery[delivery['bowler'].isin(bowler)]
run=newdf.groupby('bowler')['total_runs'].sum()
balls=newdf.groupby('bowler')['total_runs'].count()
balls=balls/6
(run/balls).sort_values().head()
```

```python
#plot a heat map for against top10 batsman and top10 bowlers according to their runs
batsman=delivery.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10).index.tolist() #getting the top 10 batsman
delivery['dismissal_kind'].value_counts()
dismissal=['caught','bowled','lbw','stumped','caught and bowled','hit wicket']
out=delivery[delivery['dismissal_kind'].isin(dismissal)]
bowler=out.groupby('bowler')['dismissal_kind'].count().sort_values(ascending=False).head(10).index.tolist() #getting the top 10 bowlers name
```

```python
batsmandf=delivery[delivery['batsman'].isin(batsman)]
finaldf=batsmandf[batsmandf['bowler'].isin(bowler)]
finaldf.shape
```

```python
pv=finaldf.pivot_table(index='bowler',columns='batsman',values='batsman_runs',aggfunc='sum')
```

```python
sns.heatmap(pv,cmap='summer')
```
